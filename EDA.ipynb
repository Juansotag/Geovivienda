```python
from selenium import webdriver  
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.by import By

opts = Options()
opts.add_argument("--start-maximized")  # ventana grande para inspeccionar
# opts.add_argument("--headless=new")   # NO headless por ahora
# Forma actualizada (Selenium 4.x)
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=opts
)
driver.get("https://www.fincaraiz.com.co/arriendo/apartamentos-y-apartaestudios/bogota/bogota-dc")
time.sleep(3)  # deja el navegador abierto

# Seleccionamos un elemento por el Xpath y lo analizamos
xpath = f"/html/body/div[2]/div[2]/div/section/div[1]/div/div/div[2]/div"
elemento = driver.find_element(By.XPATH, xpath)

def buscar_elemento(driver, xpath):
    try:
        elemento = driver.find_element(By.XPATH, xpath)
        return elemento.text
    except Exception as e:
        print(f"Error al buscar el elemento: {e}")
        return None

# Aplicamos la función
texto = buscar_elemento(driver, xpath)
print(texto)
driver.quit()
```

# Versión funcional

```python
opts = Options()
opts.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=opts
)

# Acceder a la página principal
driver.get("https://www.fincaraiz.com.co/arriendo/apartamentos-y-apartaestudios/bogota/bogota-dc")
time.sleep(3)

# Creamos una lista para almacenar los links
links_dataset = []

for j in range(1, 10):
    print(f"\n Procesando página {j}")
    consecutive_errors = 0
    max_consecutive_errors = 7
    i = 1
    while consecutive_errors < max_consecutive_errors:
        try:
            driver.execute_script("window.scrollBy(0, 500);")
            time.sleep(0.2)
            anuncio = driver.find_element(
                By.XPATH,
                f"/html/body/div[2]/div[2]/div/section/div[{i}]/div/div/div[2]/a[2]"
            )
            link = anuncio.get_attribute("href")
            print(f"Link anuncio {i}: {link}")
            
            # Add link to dataset
            links_dataset.append(link)
            
            # Reset error counter on success
            consecutive_errors = 0
            i += 1
        except Exception as e:
            print(f"Error en anuncio {i}: {e}")
            consecutive_errors += 1
            i += 1
            
            if consecutive_errors >= max_consecutive_errors:
                print(f"\n Deteniendo: {consecutive_errors} errores consecutivos")
                consecutive_errors = 0
                break
    print(f"\n Total de links recolectados: {len(links_dataset)} en la página {j}")

    # Pasamos página
    time.sleep(3)
    siguiente = driver.find_element(By.LINK_TEXT, ">")
    print("Pasando página")
    siguiente.click()
    time.sleep(7)

# Total de links recolectados
print(f"\n Total de links recolectados: {len(links_dataset)}")

# Cerramos el navegador
driver.quit()
```

# Versión experimental

```python
from selenium import webdriver  
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
from selenium.webdriver.common.by import By

opts = Options()
opts.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=opts
)

# Acceder a la página principal
driver.get("https://www.fincaraiz.com.co/arriendo/apartamentos-y-apartaestudios/bogota/bogota-dc")
time.sleep(3)

# Creamos una lista para almacenar los links
links_dataset = []

for j in range(1, 10):
    print(f"\nProcesando página {j}")
    time.sleep(2)
    
    # Buscar TODOS los anuncios usando el patrón del XPATH que funciona
    anuncios_encontrados = 0
    for i in range(1, 30):  # Intentar hasta 30 posiciones por página
        try:
            driver.execute_script("window.scrollBy(0, 300);")
            time.sleep(0.1)
            
            anuncio = driver.find_element(
                By.XPATH,
                f"/html/body/div[2]/div[2]/div/section/div[{i}]/div/div/div[2]/a[2]"
            )
            link = anuncio.get_attribute("href")
            
            if link and link not in links_dataset:
                links_dataset.append(link)
                anuncios_encontrados += 1
                print(f"Link {anuncios_encontrados}: {link}")
        except:
            # Silenciosamente continuar si no encuentra el elemento
            pass
    
    print(f"Anuncios encontrados en página {j}: {anuncios_encontrados}")
    print(f"Total acumulado: {len(links_dataset)}")
    
    # Pasar a siguiente página
    time.sleep(3)
    siguiente = driver.find_element(By.LINK_TEXT, ">")
    print("Pasando página")
    siguiente.click()
    time.sleep(7)

# Total de links recolectados
print(f"\nTotal de links recolectados: {len(links_dataset)}")

# Cerramos el navegador
driver.quit()
```
